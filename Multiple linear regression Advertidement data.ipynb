```python
y=dataset.Sales
```

```python
x=dataset.drop('Sales', axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
MReg=LinearRegression()
```

```python
MReg.fit(x_train,y_train)
```

```python
predictions=MReg.predict(x_test)
```

```python
predictions[0]
```

```python
print(y_test[0])
```

```python
r2_score(y_test,predictions)
```
